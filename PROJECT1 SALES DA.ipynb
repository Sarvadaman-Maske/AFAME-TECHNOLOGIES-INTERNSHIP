### PROJECT 1: SALES DATA ANALYSIS 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_data = pd.read_excel("C:\\OneDrive\\Desktop\\DA\\ECOMM DATA.xlsx")
```

```python
# Calculate total sales
total_sales = sales_data['Sales'].sum()
print(f"Total Sales: {total_sales}")
```

```python
# Convert order date to datetime type
sales_data['Order_Date'] = pd.to_datetime(sales_data['Order Date'], format='%d-%m-%Y')
```

```python
# Set the order date as the index
sales_data.set_index('Order_Date', inplace=True)
```

```python
# Aggregate sales by month
monthly_sales = sales_data['Sales'].resample('M').sum()

# Plot sales trends over time
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='blue')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)

# Save the figure with high resolution
plt.savefig('monthly_sales_trends.jpg', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
# Aggregate sales by product
best_selling_products = sales_data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)

print(best_selling_products.head(10))
```

```python
# Top 10 best-selling products
top_10_products = best_selling_products.head(10)

# Plot top 10 best-selling products
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_products.values, y=top_10_products.index, palette='viridis')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')

plt.savefig('top_10.jpg',dpi=300,bbox_inches='tight')
plt.show()
```

```python

sales_data['Order_Date'] = pd.to_datetime(sales_data['Order Date'], format='%d-%m-%Y')
sales_data['Ship_Date'] = pd.to_datetime(sales_data['Ship Date'], format='%d-%m-%Y')
sales_data.dropna(inplace=True)
```

```python
# Calculate total sales
total_sales = sales_data['Sales'].sum()
print(f"Total Sales: {total_sales}")

# Calculate total profit
total_profit = sales_data['Profit'].sum()
print(f"Total Profit: {total_profit}")

# Calculate average discount
average_discount = sales_data['Discount'].mean()
print(f"Average Discount: {average_discount}")

# Calculate total quantity sold
total_quantity = sales_data['Quantity'].sum()
print(f"Total Quantity Sold: {total_quantity}")
```

```python
# Aggregate sales by category
sales_by_category = sales_data.groupby('Category')['Sales'].sum()

# Plot sales by category
plt.figure(figsize=(12, 6))
sns.barplot(x=sales_by_category.values, y=sales_by_category.index, palette='viridis')
plt.title('Sales by Category')
plt.xlabel('Total Sales')
plt.ylabel('Category')
plt.savefig('salesbycat.jpg',dpi=300,bbox_inches='tight')
plt.show()
```

```python
# Aggregate profit by region
profit_by_region = sales_data.groupby('Region')['Profit'].sum()

# Plot profit by region
plt.figure(figsize=(12, 6))
sns.barplot(x=profit_by_region.values, y=profit_by_region.index, palette='viridis')
plt.title('Profit by Region')
plt.xlabel('Total Profit')
plt.ylabel('Region')
plt.savefig('profitbyreg.jpg',dpi=300,bbox_inches='tight')
plt.show()
```
